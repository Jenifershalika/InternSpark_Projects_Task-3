```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("traffic.csv")  # replace "your_file.csv" with your actual file name


print(df.head())       # shows first 5 rows
print(df.columns)      # shows all column names
print(df.info())       # checks data types and missing values
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dropna(inplace=True)
```

```python
total_events = df.shape[0]
print("Total Events:", total_events)
```

```python
unique_links = df['linkid'].nunique()
print("Unique Links:", unique_links)
```

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top Countries by Traffic")
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

```python
df['event'].value_counts()
```

```python
sns.countplot(data=df, x='event')
plt.title("Event Distribution")
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

```python
top_artists = df['artist'].value_counts().head(10)

top_artists.plot(kind='bar')
plt.title("Top Artists")
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

```python
top_tracks = df['track'].value_counts().head(10)

top_tracks.plot(kind='bar')
plt.title("Top Tracks")
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

```python
country_traffic = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=country_traffic.index, y=country_traffic.values)
plt.title("Traffic by Country")
plt.xticks(rotation=45)
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

```python
daily_traffic = df.groupby(df['date'].dt.date).size()

daily_traffic.plot()
plt.title("Daily Traffic Trend")
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

```python
top_links = df['linkid'].value_counts().head(10)

top_links.plot(kind='bar')
plt.title("Top Landing Links")
plt.show()
```

Observation:
USA contributes the highest number of traffic events,
indicating strong listener engagement from this region.

• Most traffic comes from USA and UK
• Click events dominate user interactions
• Certain artists receive significantly higher engagement
• Traffic peaks during specific dates
• Few links generate majority of user traffic
